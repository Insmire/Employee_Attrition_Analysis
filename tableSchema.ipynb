```python
# dependencies
import pandas as pd
```

```python
# read in data
file_path = "resources/IBM_employee_data.csv"
attrition_df = pd.read_csv(file_path)
attrition_df.head()
```

```python
# list of columns from dataframe
attrition_df.columns
```

```python
attrition_df.dtypes
```

PersonalFactors Table

```python
# columns in table
personal_cols = ["Age", "DistanceFromHome", "Education", "EducationField", "EmployeeNumber", 
                 "EnvironmentSatisfaction", "Gender", "MaritalStatus", "NumCompaniesWorked","Over18",
                 "RelationshipSatisfaction", "TotalWorkingYears", "WorkLifeBalance"]
```

```python
# subset original dataframe
personalFactor_df = attrition_df[personal_cols]
personalFactor_df.head()
```

```python
# number of items in personal factors dataframe
len(personalFactor_df.columns)
```

ProfessionalFactors table

```python
professional_cols = ['Attrition', 'BusinessTravel', 'DailyRate', 'Department', 'EmployeeCount',
       'EmployeeNumber', 'HourlyRate',
       'JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction',
       'MonthlyIncome', 'MonthlyRate',
       'OverTime', 'PercentSalaryHike', 'PerformanceRating',
       'StandardHours', 'StockOptionLevel', 'TrainingTimesLastYear',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']
```

```python
# subset original dataframe
professionalFactors_df = attrition_df[professional_cols]
professionalFactors_df.head()
```

```python
# number of items in professional factors dataframe
len(professionalFactors_df.columns)
```
