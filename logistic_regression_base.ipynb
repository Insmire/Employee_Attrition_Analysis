```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_path = "resources/IBM_Employee_data.csv"
attrition_df = pd.read_csv(file_path)
attrition_df.head()
```

```python
attrition_df.dtypes
```

```python
clean_df  = pd.get_dummies(data=attrition_df, columns=['EducationField', 'Department', 'BusinessTravel', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime'])
```

```python
clean_df.head()
```

```python
columns_to_drop = ["DailyRate", "HourlyRate", "MonthlyRate", "StandardHours", "EmployeeCount", "EmployeeNumber"]
```

```python
clean_df.drop(columns=columns_to_drop, inplace=True)
clean_df.dtypes
```

```python
clean_df.dtypes
```

```python
#create features and target variables
X = clean_df.drop("Attrition",1).values
y=clean_df["Attrition"] #target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=34, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```
